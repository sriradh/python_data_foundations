# Trade-offs in Machine Learning

... aka "What's the best model?"

# Major trade-off #1: Bias/Variance

ML Models have three sources of error:

- **Bias** is error introduced because the model is too inflexible (e.g. using a linear model for a nonlinear process)
- **Variance** is the error introduced because the model is too sensitive to random variations in the input data
- **Irreducible error** is error introduced because we don't have the right features (or the causes of variation are uknown)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
degree = 12
np.random.seed(1)

xs = np.random.random(degree) * 10
real = np.vectorize(lambda x: (10 - (x-4)**2))
ys = real(xs) + 2 * np.random.randn(xs.shape[0])

X = xs[:,np.newaxis]
m0 = LinearRegression()
m0.fit(X, ys)
m1 = SVR(C=10)
m1.fit(X, ys)
m2 = np.poly1d(np.polyfit(xs, ys, deg=degree))

x_pred = np.c_[:xs.max():100j]
```

## High bias: Underfitting

High bias indicates that our model simply doesn't correspond to the data (not even the _training_ data)

```python
ax = plt.axes()
ax.set(title='Underfit (high bias) model\n(LinearRegression)')
ax.scatter(xs, ys)
ax.plot(x_pred, real(x_pred), color='green', label='Actual')
ax.plot(x_pred, m0.predict(x_pred), color='red', label='Model')
ax.legend();
```

## High variance: Overfitting

High variance indicates that we've learned the training data *really well*, but our model doesn't *generalize*

```python
ax = plt.axes()
ax.set(title='Overfit (high variance) model\n(Polynomial)')
ax.scatter(xs, ys)
ax.plot(x_pred, real(x_pred), color='green', label='Actual')
ax.plot(x_pred, m2(x_pred), color='red', label='Model')
ax.legend();
```

# Balanced model

What we need to do is balance bias vs. variance

```python
ax = plt.axes()
ax.set(title='Good model (low bias, low variance)\n(SVM)')
ax.scatter(xs, ys)
ax.plot(x_pred, real(x_pred), color='green', label='Actual')
ax.plot(x_pred, m1.predict(x_pred), color='red', label='Model')
ax.legend();
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(2, 2, figsize=(16, 12))
ax0.set(title='Underfit (high bias) model\n(LinearRegression)')
ax0.scatter(xs, ys)
ax0.plot(x_pred, real(x_pred), color='green', label='Actual')
ax0.plot(x_pred, m0.predict(x_pred), color='red', label='Model')
ax0.legend();

ax1.set(title='Overfit (high variance) model\n(Polynomial)')
ax1.scatter(xs, ys)
ax1.plot(x_pred, real(x_pred), color='green', label='Actual')
ax1.plot(x_pred, m2(x_pred), color='red', label='Model')
ax1.legend();

ax2.set(title='Good model (low bias, low variance)\n(SVM)')
ax2.scatter(xs, ys)
ax2.plot(x_pred, real(x_pred), color='green', label='Actual')
ax2.plot(x_pred, m1.predict(x_pred), color='red', label='Model')
ax2.legend();
```

```python
xs = np.r_[0:1:50j]
pwr = 6
var = np.exp(xs) ** 6
bias = np.exp(1 - xs) ** 6
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(xs, var, label='Variance')
ax.plot(xs, bias, label='Bias')
ax.plot(xs, var + bias + 100, label='Total error')
i = len(xs)//2
ax.plot(xs[i], var[i] + bias[i] + 100, ms=20, marker='x', label='Best', linestyle='none')
ax.legend();
```

# Flexibility / sensitivity / power as a source of error

The more *flexible* our model, the more prone to overfitting (and having high variance)

# Major trade-off #2: Interpretability vs power / accuracy

The most flexible/powerful models (gradient boosting trees, SVM, neural networks, etc.) are also the most difficult to explain

Explanatory power can motivate **human decision-makers**, so often a weaker model is a "better" solution because it's easy to understand

Sometimes the **true** objective of your model is to "make sure I don't look foolish."

# Major trade-off #3: Power / accuracy vs performance constraints

The most powerful model ...

... that needs to be deployed to a mobile phone

... and requires 2 TB of RAM for inference
