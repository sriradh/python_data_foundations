# Principal Component Analysis

In some cases, one-hot encoding and other feature engineering can result in an explosion of features.

The increase in the number of features can pose problems with computational tractability and with accuracy (more features is not always better).

PCA provides us a way to reduce the number of features (dimensions) without losing *too* much data:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.random.random(100) * 10
y = 0.5 * x - 2 + np.random.randn(100)
X = np.c_[x, y]

ax = plt.axes()
ax.scatter(X[:,0], X[:,1], alpha=0.5)
ax.set_aspect('equal', adjustable='datalim')
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
ax = plt.axes()
ax.scatter(X[:,0], X[:,1], alpha=0.5)
ax.set_aspect('equal', adjustable='datalim')
arrowprops=dict(arrowstyle='->',
                linewidth=2,
                shrinkA=0, shrinkB=0)

ax.annotate(
    '', 
    pca.mean_ + pca.components_[0] * np.sqrt(pca.explained_variance_ratio_[0]) * 5, 
    pca.mean_, 
    arrowprops=arrowprops
)
ax.annotate(
    '', 
    pca.mean_ + pca.components_[1] * np.sqrt(pca.explained_variance_ratio_[1]) * 5, 
    pca.mean_, 
    arrowprops=arrowprops
);
```

```python
X1 = pca.transform(X)
ax = plt.axes()
ax.scatter(X1[:,0], X1[:,1])
ax.axis('equal');
```

```python
pca2 = PCA(n_components=1)
Xt = pca2.fit_transform(X)
X2 = pca2.inverse_transform(Xt)

ax = plt.axes()
ax.scatter(X[:,0], X[:,1], alpha=0.5, color='b')
ax.scatter(X2[:,0], X2[:,1], alpha=0.5, color='r')
```

Let's go to higher dimensions:

```python
from sklearn.datasets import load_wine

wine = load_wine()
wine.feature_names
```

```python
wine.target_names
```

```python
pca = PCA()
pca.fit(wine.data)
pca.explained_variance_ratio_
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
import pandas as pd
pd.DataFrame(wine.data, columns=wine.feature_names).describe()
```

```python
pca = PCA(n_components=1)
wine_1 = pca.fit_transform(wine.data)
plt.scatter(wine_1, wine.target)
```

# Let's see how accuracy changes with different #s of dimensions for wine in a decision tree

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
dims = np.r_[1: len(wine.feature_names) + 1]
accuracies = []

for dim in dims:
    pca = PCA(n_components=dim)
    X = pca.fit_transform(wine.data)
    X_train, X_test, y_train, y_test = train_test_split(X, wine.target, random_state=1)
    model = DecisionTreeClassifier()
    model.fit(X_train, y_train)
    accuracies.append(accuracy_score(y_test, model.predict(X_test)))
```

```python
plt.plot(dims, accuracies)
```

# Scaling our inputs

It turns out that since PCA is a *geometric* approach, we can get much better results if we scale our data using a StandardScaler:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(wine.data)
```

```python
dims = np.r_[1: len(wine.feature_names) + 1]
accuracies = []

for dim in dims:
    pca = PCA(n_components=dim)
    X = pca.fit_transform(scaled_data)
    X_train, X_test, y_train, y_test = train_test_split(X, wine.target, random_state=1)
    model = DecisionTreeClassifier()
    model.fit(X_train, y_train)
    accuracies.append(accuracy_score(y_test, model.predict(X_test)))
```

```python
plt.plot(dims, accuracies);
```

```python
pca = PCA(n_components=2)
X0 = pca.fit_transform(wine.data)
X1 = pca.fit_transform(scaled_data)
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(12,6))

ax0.set(title='Unscaled data', xlabel='component0', ylabel='component1')
ax0.scatter(X0[:,0], X0[:,1], c=wine.target)

ax1.set(title='Scaled data', xlabel='component0', ylabel='component1')
ax1.scatter(X1[:,0], X1[:,1], c=wine.target);
```

```python
pca_full = PCA(0.95)
pca_full.fit(scaled_data)
plt.plot(np.cumsum(pca_full.explained_variance_ratio_))
```
