# Support Vector Machines (SVM)

SVM can be used for both *classification* and *regression* but is easiest to explain for *classification*.

Let's consider some 'blob' data:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(centers=2, cluster_std=2)
```

```python
plt.scatter(X[:,0], X[:,1], c=y);
```

## SVM big ideas

- Divide input samples into categories by drawing a straight line (if we have 2-D data) or a (hyper-)plane for higher-dimensional data.
- Choose that line to 'maximize the margin' (keep samples as far as possible from the dividing line)

To do this, it finds relevant input samples (close to the decision boundary) and ignores everything else. The relevant samples are the "support vectors"



# SVM in action

Let's create a support vector classifier (SVC) with default arguments:

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear')
```

```python
svc.fit(X, y)
```

We can find the support vectors the model chose and plot them on our graph:

```python
svc.support_vectors_
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(svc.support_vectors_[:,0], svc.support_vectors_[:, 1], marker='x', s=100)
```

We can also visualize the decision boundary by making predictions for a grid of points:

```python
## Code from https://scikit-learn.org/stable/auto_examples/svm/plot_iris_svc.html

def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
xx, yy = make_meshgrid(X[:,0], X[:,1])

ax = plt.axes()
plot_contours(ax, svc, xx, yy, alpha=0.1)
ax.scatter(X[:,0], X[:,1], c=y)
ax.scatter(svc.support_vectors_[:,0], svc.support_vectors_[:, 1], marker='x', s=100)
```

# SVM Kernels

Sometimes, a straight/flat decision boundary is not what we want. In this case, we can transform our input data into a higher # of dimensions before actually dividing it. 

This is called the "kernel trick". Consider the following dataset:

```python
from sklearn.datasets import make_circles

X, y = make_circles(factor=0.2, noise=0.1, random_state=1)
ax = plt.axes()
ax.scatter(X[:,0], X[:,1], c=y)
```

A linear SVC won't do a great job:

```python
svc = SVC(kernel='linear')
svc.fit(X, y)
```

```python
xx, yy = make_meshgrid(X[:,0], X[:,1])

ax = plt.axes()
plot_contours(ax, svc, xx, yy, alpha=0.1)
ax.scatter(X[:,0], X[:,1], c=y)
ax.scatter(svc.support_vectors_[:,0], svc.support_vectors_[:, 1], marker='x', s=100)
```

We could, however, add a third dimension based on the distance from the center (a 'radial basis function'):

```python
X3 = X[:,0] ** 2 + X[:,1] ** 2
```

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], X3, c=y)
```

SVC lets us do this by specifying 'rbf' for the kernel:

```python
svc = SVC(kernel='rbf')
svc.fit(X, y)
```

```python
xx, yy = make_meshgrid(X[:,0], X[:,1])

ax = plt.axes()
plot_contours(ax, svc, xx, yy, alpha=0.1)
ax.scatter(X[:,0], X[:,1], c=y)
ax.scatter(svc.support_vectors_[:,0], svc.support_vectors_[:, 1], marker='x', s=100)
```

# Multi-class classification

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data
y = iris.target

X = StandardScaler().fit_transform(X)
X = PCA(n_components=2).fit_transform(X)

ax = plt.axes()
ax.scatter(X[:,0], X[:,1], c=y);
```

```python
svc_iris = SVC(kernel='linear')
svc_iris.fit(X, y)
```

```python
xx, yy = make_meshgrid(X[:,0], X[:,1])

ax = plt.axes()
plot_contours(ax, svc_iris, xx, yy, alpha=0.1)
ax.scatter(X[:,0], X[:,1], c=y)
ax.scatter(svc_iris.support_vectors_[:,0], svc_iris.support_vectors_[:, 1], marker='x', s=100)
```

```python
wine = datasets.load_wine()

X = wine.data
y = wine.target


X = StandardScaler().fit_transform(X)
X = PCA(n_components=2).fit_transform(X)
```

```python
clf = SVC(kernel='rbf')
clf.fit(X, y)
```

```python
xx, yy = make_meshgrid(X[:,0], X[:,1])

ax = plt.axes()
plot_contours(ax, clf, xx, yy, alpha=0.1)
ax.scatter(X[:,0], X[:,1], c=y)
ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:, 1], marker='x', s=100)
```

# Conclusion and Caveats

- SVM is a geometry-based algorithm, so it's sensitive to scaling issues. Ensure that you have pre-scaled your input before using SVM
- SVM has both classification (SVC) and regression (SVR) forms in sklearn
- SVM can be expensive to compute
- SVM is frequently difficult to interpret ("clusters together in a high-dimensional vector space") so may not work for some applications
