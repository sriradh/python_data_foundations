# Random variables and distributions

A **random variable** is a variable whose value depends on a random phenomenon.

A **random distribution** is the probability that a random variable will take on its various possible values


# Uniform random variables

Takes on a value in a range with equal probability (usually 0..1)

```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
values = np.random.random(1000)
ax = plt.axes()
sns.distplot(values, ax=ax)
ax.set(title='Uniform random distribution');
```

## Gaussian ('normal') random variables

Characterized by a mean (center) and a standard deviation

Useful for modelling physical variation (the height of a person is normally distributed)

```python
values = np.random.normal(loc=3, scale=2, size=1000)
ax = plt.axes()
sns.distplot(values, ax=ax)
ax.set(title='Gaussian distribution (loc=3, scale=2)');
```

One nice thing about gaussians is that we can predict the % of the time we get certain values based on how far away from the mean they are:

- mean +/- one standard deviation: 68%
- mean +/- two standard deviations: 95%
- mean +/- three standard deviations: 99.7%

## Poisson random variables

Poisson variables generally model a process where there is a probability of an event happening in a certain time period (and that there is no dependency from one time period to another).

You could use it to model the number of sales orders on a given day, for example.

It is characterized by the mean # of events in a time period ('lambda')

```python
values = np.random.poisson(5, 1000)
ax = plt.axes()
sns.distplot(values, ax=ax)
ax.set(title='Poisson distribution (lambda=5)');
```

## Chi-Squared

This RV is actually the sum of `k` independent gaussian random variables with mean=0 and std=1, squared. 

```python
ax = plt.axes()
ax.set(title='Chi-Squared distributions')
colors='rgbcmy'
for k in range(1,7):
    values = np.random.chisquare(k, 1000)
    sns.distplot(values, ax=ax, color=colors[k-1], hist=False, label=f'k={k}')
ax.legend();
```

## Binomial

Models a binary-outcome test (coin flip, for example)

The variable models the # of times the event occurred

```python
values = np.random.binomial(1000, 0.5, 1000)  # 1000 coin flips
ax = plt.axes()
sns.distplot(values, ax=ax)
ax.set(title='Binomial distribution (p=0.5)');
```

## Exponential

Exponential random variables track the amount of time between events in a poisson process

```python
values = np.random.exponential(5, size=1000)
ax = plt.axes()
sns.distplot(values, ax=ax)
ax.set(title='Exponential distribution (scale=5)')
```
